```python
import pandas as pd

df_students = pd.read_csv('students.csv')
print(df_students.shape)
df_feedback = pd.read_csv('feedback.csv')
print(df_feedback.shape)
df_course_activity = pd.read_csv('course_activity.csv')
print(df_course_activity.shape)
```

```python
dataframes = [df_students, df_feedback, df_course_activity]
dataframe_names = ["df_students", "df_feedback", "df_course_activity"]

for df, name in zip(dataframes, dataframe_names):
    print(f"Exploring {name}:")
    display(df.head())
    print(df.info())
    display(df.describe(include='all'))
    print("Missing values per column:\n", df.isnull().sum())
    print("Number of duplicate rows:", df.duplicated().sum())
    print("-" * 50)
```

```python
#Conversion to datetime format
try:
    df_students['Enrolment_Date'] = pd.to_datetime(df_students['Enrolment_Date'], format='%d-%m-%Y', errors='raise')
except ValueError as e:
    print(f"Error converting 'Enrolment_Date': {e}")

try:
    df_course_activity['Date'] = pd.to_datetime(df_course_activity['Date'], format='%d-%m-%Y', errors='raise')
except ValueError as e:
    print(f"Error converting 'Date': {e}")


if (df_students['Age'] < 18).any() or (df_students['Age'] > 100).any():
    print("Unusual age values found. Further investigation needed.")

for col in ['Gender', 'Location']:
    print(f"Unique values in '{col}': {df_students[col].unique()}")

#Checking of rating less than 1 or greater than 5
if (df_feedback['Rating'] < 1).any() or (df_feedback['Rating'] > 5).any():
    print("Rating values outside of 1-5 found. Further investigation needed.")

#Checking of time spent less than 0 minutes
if (df_course_activity['Time_Spent_Minutes'] < 0).any():
    print("Negative time spent values found. Further investigation needed.")

#Completion percentage should be within 0-100
if (df_course_activity['Completion_Percentage'] < 0).any() or (df_course_activity['Completion_Percentage'] > 100).any():
    print("Completion percentage outside 0-100 found. Further investigation needed.")

display(df_students.head())
display(df_feedback.head())
display(df_course_activity.head())
```

```python
#What is the overall average completion rate across courses?

#DM101 completion rate
average_completion_rate_DM101=df_course_activity[df_course_activity['Course_ID'] == 'DM101']['Completion_Percentage'].mean()
print(f"The average completion rate for DM101 is: {average_completion_rate_DM101:.2f}%")
#PY202 completion rate
average_completion_rate_PY202=df_course_activity[df_course_activity['Course_ID'] == 'PY202']['Completion_Percentage'].mean()
print(f"The average completion rate for PY202 is: {average_completion_rate_PY202:.2f}%")
#UX303 completion rate
average_completion_rate_UX303=df_course_activity[df_course_activity['Course_ID'] == 'UX303']['Completion_Percentage'].mean()
print(f"The average completion rate for UX303 is: {average_completion_rate_UX303:.2f}%")
#WD404 completion rate
average_completion_rate_WD404=df_course_activity[df_course_activity['Course_ID'] == 'WD404']['Completion_Percentage'].mean()
print(f"The average completion rate for WD404 is: {average_completion_rate_WD404:.2f}%")
```

```python
#Which course has the highest and lowest average engagement time?
average_engagement_DM101=df_course_activity[df_course_activity['Course_ID'] == 'DM101']['Time_Spent_Minutes'].mean()
average_engagement_PY202=df_course_activity[df_course_activity['Course_ID'] == 'PY202']['Time_Spent_Minutes'].mean()
average_engagement_UX303=df_course_activity[df_course_activity['Course_ID'] == 'UX303']['Time_Spent_Minutes'].mean()
average_engagement_WD404=df_course_activity[df_course_activity['Course_ID'] == 'WD404']['Time_Spent_Minutes'].mean()

engagement_times = {
    'DM101': average_engagement_DM101,
    'PY202': average_engagement_PY202,
    'UX303': average_engagement_UX303,
    'WD404': average_engagement_WD404
}

max_engagement_course = max(engagement_times, key=engagement_times.get)
min_engagement_course = min(engagement_times, key=engagement_times.get)

print(f"The course with the highest average engagement time is {max_engagement_course} with {engagement_times[max_engagement_course]:.2f} minutes.")
print(f"The course with the lowest average engagement time is {min_engagement_course} with {engagement_times[min_engagement_course]:.2f} minutes.")
```

```python
#How does engagement differ by age group or location?

# Merge the necessary dataframes
merged_df = pd.merge(df_students, df_course_activity, on='Student_ID', how='inner')

# Analyze engagement by age group
age_groups = pd.cut(merged_df['Age'], bins=[18, 25, 30, 35], labels=['18-24', '25-30', '35-40'])
engagement_by_age = merged_df.groupby(age_groups)['Time_Spent_Minutes'].mean()
print("Average Engagement Time by Age Group:\n", engagement_by_age)

# Analyze engagement by location
engagement_by_location = merged_df.groupby('Location')['Time_Spent_Minutes'].mean()
print("\nAverage Engagement Time by Location:\n", engagement_by_location)
```

```python
#What is the average feedback rating per course?
average_feedback_rating_DM101=df_feedback[df_feedback['Course_ID'] == 'DM101']['Rating'].mean()
print(f"The average feeback rate for DM101 is: {average_feedback_rating_DM101:.2f}/5")
average_feedback_rating_PY202=df_feedback[df_feedback['Course_ID'] == 'PY202']['Rating'].mean()
print(f"The average feeback rate for PY202 is: {average_feedback_rating_PY202:.2f}/5")
average_feedback_rating_UX303=df_feedback[df_feedback['Course_ID'] == 'UX303']['Rating'].mean()
print(f"The average feeback rate for UX303 is: {average_feedback_rating_UX303:.2f}/5")
average_feedback_rating_WD404=df_feedback[df_feedback['Course_ID'] == 'WD404']['Rating'].mean()
print(f"The average feeback rate for WD404 is: {average_feedback_rating_WD404:.2f}/5")
```

```python
#Is there a correlation between completion rate and feedback rating?

# Merge the course activity and feedback dataframes
merged_df = pd.merge(df_course_activity, df_feedback, on=['Student_ID', 'Course_ID'], how='inner')

# Calculate the correlation between completion rate and feedback rating
correlation = merged_df['Completion_Percentage'].corr(merged_df['Rating'])
if(correlation > 0):
    print("There is a positive correlation between completion rate and feedback rating.")
elif(correlation < 0):
    print("There is a negative correlation between completion rate and feedback rating.")
else:
    print("There is no correlation between completion rate and feedback rating.")

print(f"The correlation between completion rate and feedback rating is: {correlation:.2f}")
```

```python
#Identify top 3 student segments based on engagement and satisfaction.

# Calculate a combined engagement and satisfaction score
merged_df['Engagement_Score'] = merged_df['Time_Spent_Minutes'] * merged_df['Completion_Percentage'] / 100
merged_df['Combined_Score'] = merged_df['Engagement_Score'] * merged_df['Rating']

# Group by student and get the mean combined score
student_segments = merged_df.groupby('Student_ID')['Combined_Score'].mean().sort_values(ascending=False)

# Get the top 3 student segments
top_3_segments = student_segments.head(3)

print("Top 3 Student Segments based on Combined Engagement and Satisfaction:")
top_3_segments
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Engagement patterns by course
plt.figure(figsize=(10, 6))
sns.barplot(x='Course_ID', y='Time_Spent_Minutes', data=df_course_activity)
plt.title('Average Engagement Time by Course')
plt.xlabel('Course ID')
plt.ylabel('Average Time Spent (Minutes)')
plt.show()
```

```python
# 2. Engagement patterns by age group
plt.figure(figsize=(10, 6))
sns.barplot(x=age_groups, y='Time_Spent_Minutes', data=merged_df)
plt.title('Average Engagement Time by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Time Spent (Minutes)')
plt.show()
```

```python
# 3. Feedback ratings per course
plt.figure(figsize=(10, 6))
course_feedback = df_feedback.groupby('Course_ID')['Rating'].mean().reset_index()
sns.barplot(x='Course_ID', y='Rating', data=course_feedback)
plt.title('Average Feedback Rating by Course')
plt.xlabel('Course ID')
plt.ylabel('Average Rating')
plt.ylim(0, 5)
plt.show()
```

```python
# 4. Completion rate vs. feedback rating (Scatter Plot)
plt.figure(figsize=(8, 6))
sns.regplot(x='Completion_Percentage', y='Rating', data=merged_df)
plt.title('Completion Rate vs. Feedback Rating with Regression')
plt.xlabel('Completion Percentage')
plt.ylabel('Feedback Rating')
plt.show()
```

```python
completion_over_time = df_course_activity.groupby('Date')['Completion_Percentage'].mean().reset_index()
 # Create the line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Completion_Percentage', data=completion_over_time)
plt.title('Trend of Average Completion Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Average Completion Percentage')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
engagement_over_time = df_course_activity.groupby('Date')['Time_Spent_Minutes'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Time_Spent_Minutes', data=engagement_over_time)
plt.title('Trend of Average Engagement Time Over Time')
plt.xlabel('Date')
plt.ylabel('Average Time Spent (Minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
